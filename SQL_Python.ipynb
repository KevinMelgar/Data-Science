## Problems

Now write and execute SQL queries to solve assignment problems

### Problem 1

##### Find the total number of crimes recorded in the CRIME table.


```python
%%sql select count(CASE_NUMBER) as ' Number of crimes'
from CHICAGO_CRIME ;
```

### Problem 2

##### List community area names and numbers with per capita income less than 11000.


```python
%%sql select COMMUNITY_AREA_NAME, COMMUNITY_AREA_NUMBER
from CHICAGO_CENSUS
where PER_CAPITA_INCOME > 11000 order by PER_CAPITA_INCOME NULLS LAST ;
```

### Problem 3

##### List all case numbers for crimes involving minors?(children are not considered minors for the purposes of crime analysis)


```python
%%sql select CASE_NUMBER, DESCRIPTION, PRIMARY_TYPE
from CHICAGO_CRIME
where DESCRIPTION like '%MINOR%' or PRIMARY_TYPE like '%MINOR%';
```

### Problem 4

##### List all kidnapping crimes involving a child?


```python
%%sql select CASE_NUMBER, DESCRIPTION, PRIMARY_TYPE
from CHICAGO_CRIME
where DESCRIPTION like '%CHILD%' and PRIMARY_TYPE like '%KIDNAPPING%';
```

### Problem 5

##### List the kind of crimes that were recorded at schools. (No repetitions)


```python
%%sql select distinct LOCATION_DESCRIPTION, DESCRIPTION
from CHICAGO_CRIME
WHERE LOCATION_DESCRIPTION LIKE '%SCHOOL%' OR LOCATION_DESCRIPTION LIKE '%SCHOOLS%';
```

```python
%%sql select CASE_NUMBER, LOCATION_DESCRIPTION, DESCRIPTION, PRIMARY_TYPE
from CHICAGO_CRIME
WHERE LOCATION_DESCRIPTION LIKE '%SCHOOL%' OR LOCATION_DESCRIPTION LIKE '%SCHOOLS%';
```

### Problem 6

##### List the type of schools along with the average safety score for each type.


```python
%%sql
SELECT
    CASE
        WHEN "Elementary, Middle, or High School" = 'ES' THEN 'Elementary School'
        WHEN "Elementary, Middle, or High School" = 'MS' THEN 'Middle School'
        WHEN "Elementary, Middle, or High School" = 'HS' THEN 'High School'
        ELSE 'Unknown'
    END AS School_Type,
    AVG(SAFETY_SCORE) AS Average_Safety_Score
FROM CHICAGO_SHOOLS
GROUP BY School_Type;
```

### Problem 7

##### List 5 community areas with highest % of households below poverty line


```python
%%sql select MAX(COMMUNITY_AREA_NAME), PERCENT_HOUSEHOLDS_BELOW_POVERTY as "Highest % of households below poverty line"
from CHICAGO_CENSUS
group by PERCENT_HOUSEHOLDS_BELOW_POVERTY order by PERCENT_HOUSEHOLDS_BELOW_POVERTY desc limit 5;
```

```python
%%sql
SELECT COMMUNITY_AREA_NAME, PERCENT_HOUSEHOLDS_BELOW_POVERTY as "Highest % of households below poverty line"
FROM CHICAGO_CENSUS
ORDER BY PERCENT_HOUSEHOLDS_BELOW_POVERTY DESC
LIMIT 5;
```

### Problem 8

##### Which community area is most crime prone? Display the coumminty area number only.


```python
%%sql select COMMUNITY_AREA_NUMBER as Community_Number
from CHICAGO_CRIME
group by COMMUNITY_AREA_NUMBER

ORDER BY
    COUNT(COMMUNITY_AREA_NUMBER) DESC NULLS LAST
LIMIT 1;
```

### Problem 9

##### Use a sub-query to find the name of the community area with highest hardship index


```python
%%sql
select COMMUNITY_AREA_NAME as 'Community area with highest hardship index'
from CHICAGO_CENSUS
where HARDSHIP_INDEX = (select MAX(HARDSHIP_INDEX) from CHICAGO_CENSUS);
```

### Problem 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
%%sql
SELECT
    COMMUNITY_AREA_NAME
FROM
    CHICAGO_CENSUS
WHERE
    COMMUNITY_AREA_NUMBER = (
        SELECT
            COMMUNITY_AREA_NUMBER
        FROM
            CHICAGO_CRIME
        GROUP BY
            COMMUNITY_AREA_NUMBER
        ORDER BY
            COUNT(*) DESC NULLS LAST
        LIMIT 1
    );
```
